```python
## Generate Traits

TOTAL_IMAGES = 10000 # Number of random unique images we want to generate

all_images = [] 

# A recursive function to generate unique image combinations
def create_new_image():
    
    new_image = {} 

    # For each trait category, select a random trait based on the weightings
    new_image ["Background"] = random.choices(background, background_weights)[0]
    new_image ["Livery"] = random.choices(livery, livery_weights)[0]
    new_image ["Backbling"] = random.choices(backbling, backbling_weights)[0]
    new_image ["Skeleton"] = random.choices(skeleton, skeleton_weights)[0]
    new_image ["Neck"] = random.choices(neck, neck_weights)[0]
    new_image ["Clothing"] = random.choices(clothing, clothing_weights)[0]
    new_image ["Earring"] = random.choices(earrings, earrings_weights)[0]
    new_image ["Facewear Accessories"] = random.choices(facewear_accessories, facewear_accessories_weights)[0]
    new_image ["Headwear Below"] = random.choices(headwear_below, headwear_below_weights)[0]
    new_image ["Facewear"] = random.choices(facewear, facewear_weights)[0]
    new_image ["Eyes"] = random.choices(eyes, eyes_weights)[0]
    new_image ["Eyewear"] = random.choices(eyewear, eyewear_weights)[0]
    new_image ["Headwear Above"] = random.choices(headwear_above, headwear_above_weights)[0]
    

    if new_image in all_images:
        return create_new_image()
    else:
        return new_image
    
    
# Generate the unique combinations based on trait weightings
for i in range(TOTAL_IMAGES): 
    
    new_trait_image = create_new_image()
    
    all_images.append(new_trait_image)
    
```

```python
# Returns true if all images are unique
def all_images_unique(all_images):
    seen = list()
    return not any(i in seen or seen.append(i) for i in all_images)

print("Are all images unique?", all_images_unique(all_images))
```

```python
#TODO: pick weights for all - must add up to 100
# Each image is made up a series of traits
# The weightings for each trait drive the rarity and add up to 100%

plant = ["Prickly Pear", "Snake Plant", "Zebra Cactus", "Parodia", "Golden Barrel", "Saguaro", "Echeveria"] 
plant_weights = [6, 13, 15, 18, 23, 20, 5]

pot = ["Teal", "Sunset", "Coral", "Ocean", "Red","Lime", "Blues", "Sour", "Yellow", "Purple"] 
pot_weights = [8,3,4,5,17,14,12,5,17,15]

face = ["Cute", "Side-eye", "Pretty", "Scared", "Kiss", "Grin", "Derpy"] 
face_weights = [28, 18, 19, 13, 12, 6 , 4]

background = ["Bespin", "Daylight", "Rain", "Shooting Star", "Starry Night", "Sunshine", "Sunset"] 
background_weights = [6, 19, 14, 8, 14, 19, 20]

pet = ["Bee", "Worm", "Butterfly", "Ladybug", "Spider", "Beetle", "Dragonfly", "None"] 
pet_weights = [15, 17, 8, 11, 5, 7, 4, 33]

statue = ["Dinosaur", "Gnome", "Alarm Clock", "Alien", "Mushroom", "Sandcastle", "Radio", "None"] 
statue_weights = [3, 2, 13, 4, 6, 13, 10, 49]

buddy = ["String of Pearls", "String of Nickles", "String of Hearts", "English Ivy", "Hindu Rope", "Burro's Tail", "Lamb's Tail", "None"] 
buddy_weights = [6, 15, 1, 2, 9, 5, 4, 58]

# Dictionary variable for each trait. 
# Each trait corresponds to its file name


plant_files = {
    "Prickly Pear": "prickly",
    "Snake Plant": "snake",
    "Zebra Cactus": "zebra",
    "Parodia": "parodia",
    "Golden Barrel": "barrel",
    "Saguaro": "saguaro",
    "Echeveria": "echeveria"
}

pot_files ={
    "Teal": "teal",
    "Sunset": "sunset",
    "Red": "red",
    "Coral": "coral",
    "Lime": "lime",
    "Ocean": "ocean",
    #"Cotton-candy": "pot5-01",
    "Blues": "blues",
    "Sour": "sour",
    "Yellow": "yellow",
    "Purple": "purple",

}

face_files ={
    "Cute": "cute",
    "Side-eye": "side", 
    "Pretty": "pretty", 
    "Scared": "scared", 
    "Kiss": "kiss", 
    "Grin": "grin", 
    "Derpy": "derpy"
}

background_files = {
    "Bespin": "bespin",
    "Daylight": "daylight",
    "Rain": "rain",
    "Shooting Star": "shooting",
    "Starry Night": "stars",
    "Sunshine": "sun",
    "Sunset": "sunset"
}

pet_files = {
    "Bee": "bee",
    "Worm": "worm",
    "Butterfly": "butterfly",
    "Ladybug": "ladybug",
    "Spider": "spider",
    "Beetle": "beetle",
    "Dragonfly": "dragonfly",
    "None": "none"         
}
statue_files = {
    "Dinosaur": "dino",
    "Gnome": "gnome",
    "Sandcastle": "sandcastle",
    "Alien": "alien",
    "Mushroom": "mushroom",
    "Alarm Clock": "alarm",
    "Radio": "radio",
    "None": "none"
}
buddy_files = {
    "String of Pearls": "pearls",
    "String of Nickles": "nickles", 
    "String of Hearts": "hearts", 
    "English Ivy": "ivy", 
    "Hindu Rope": "wax", 
    "Burro's Tail": "burro", 
    "Lamb's Tail": "lamb", 
    "None": "none"
}
```

```python
## Generate Traits

TOTAL_IMAGES = 6000 # Number of random unique images we want to generate

all_images = [] 

# A recursive function to generate unique image combinations
def create_new_image():
    
    new_image = {} 

    # For each trait category, select a random trait based on the weightings 
    new_image ["Plant"] = random.choices(plant, plant_weights)[0]
    new_image ["Pot"] = random.choices(pot, pot_weights)[0]
    new_image ["Face"] = random.choices(face, face_weights)[0]
    new_image ["Background"] = random.choices(background, background_weights)[0]
    new_image ["Pet"] = random.choices(pet, pet_weights)[0]
    new_image ["Statue"] = random.choices(statue, statue_weights)[0]
    new_image ["Buddy"] = random.choices(buddy, buddy_weights)[0]
    
    if new_image in all_images:
        return create_new_image()
    else:
        return new_image
    
    
# Generate the unique combinations based on trait weightings
for i in range(TOTAL_IMAGES): 
    
    new_trait_image = create_new_image()
    
    all_images.append(new_trait_image)
    
```

```python
# Returns true if all images are unique
def all_images_unique(all_images):
    seen = list()
    return not any(i in seen or seen.append(i) for i in all_images)

print("Are all images unique?", all_images_unique(all_images))
```

```python
# Add token Id to each image
i = 0
for item in all_images:
    item["tokenId"] = i
    i = i + 1
```

```python
# Get Trait Counts
background_count = {}
for item in background:
    background_count[item] = 0
    
plant_count = {}
for item in plant:
    plant_count[item] = 0

pot_count = {}
for item in pot:
    pot_count[item] = 0

pet_count = {}
for item in pet:
    pet_count[item] = 0

face_count = {}
for item in face:
    face_count[item] = 0

statue_count = {}
for item in statue:
    statue_count[item] = 0

buddy_count = {}
for item in buddy:
    buddy_count[item] = 0

for image in all_images:
    background_count[image["Background"]] += 1
    plant_count[image["Plant"]] += 1
    pot_count[image["Pot"]] += 1
    pet_count[image["Pet"]] += 1
    face_count[image["Face"]] += 1
    statue_count[image["Statue"]] += 1
    buddy_count[image["Buddy"]] += 1
    
print(background_count)
print(plant_count)
print(pot_count)
print(pet_count)
print(face_count)
print(statue_count)
print(buddy_count)
```

```python
#### Generate Metadata for all Traits 
METADATA_FILE_NAME = './metadata/all-traits.json'; 
with open(METADATA_FILE_NAME, 'w') as outfile:
    json.dump(all_images, outfile, indent=4)
```
